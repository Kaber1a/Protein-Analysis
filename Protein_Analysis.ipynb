DATA CLEANING


The data set may be made of errors such as duplicates and missing values which if analysis is made without handling them appropriately the analysis won't be accurate thus misleading . Therefore there is need to clean the data before further analysis.

```python
numeric_fp.isnull().sum()
```

```python
numeric_fp.duplicated().sum()
```

```python

z_scores = numeric_fp.apply(lambda x: (x - x.mean()) / x.std())

threshold = 3

outliers = (z_scores > threshold).sum()

outliers
```

his shows how different variables co-rrelate. For example White meat and Eggs co-orelate by 62%. From this, we can conclude that White Meat and Eggs are linearly related. If the store area increases the house price increases and vice versa.

The correlation matrix is vital as it helps in visualising a heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

This is a representation of the co-orelation matrix visually. Co-relation above 0.5 (50%) , highlights that the 2 variables strongly relate with each other and vice versa. It enables us to graphically interpret how different varibles correlate with each other and what serves as a crucial factor in decision making. For example, milk and white meat having a correlation of 62% shows that protein content found in milk and white meat is almost the same.

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numeric_fp:
    plt.figure(figsize=(8, 6))
    values=numeric_fp[col]
    plt.hist(numeric_fp [col],bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

I have used a histogram to visually represent protein content of each product in various countries. The histogram also places countries having the same protein content of each product in the same bar thus easy interpretation can be made.
